### Task 1. Create a Python script that generates a sine wave with a frequency of 10 Hz and a duration of 2 seconds. Plot the waveform using the matplotlib library.

```python
# Suggestion: Use the numpy function np.linspace() to generate the time samples, and np.sin() to generate the sine wave.
# Be sure to use a sampling frequency that follows Nyquist-Shannon Sampling Theorem
# Then use plt.plot() to plot the waveform.
# Remember to label your plot axes

# Define the sine wave parameters
frequency = 10  # Frequency of the sine wave in Hz
duration = 2  # Duration of the sine wave in seconds
sampling_frequency = 44100  # Using a common audio sampling frequency for high-quality representation

# Generate time samples
t = np.linspace(0, duration, int(sampling_frequency * duration), endpoint=False)

# Generate the sine wave
y = np.sin(2 * np.pi * frequency * t)

# Plot the waveform
plt.figure(figsize=(10, 4))
plt.plot(t, y)
plt.title('Sine Wave')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

### Task 2. Create a Python script that applies a low-pass filter to the sine wave generated in task 1. The cutoff frequency of the low-pass filter should be 5 Hz. Plot the filtered waveform using the matplotlib library.



```python
# Use a Buttworth filter for your lowpass filter. ex. b, a = sig.butter(Order, cutoff, btype = 'low', fs = sampleFrequency)
# Order input argument can be between 1-300 (ideally 1-10)
# Documentation for sig.butter: https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.butter.html
# Use sig.lfilt() to filter the sine wave. ex. y = sig.lfilt(b, a, x)
# Documentation for sig.lfilt(): https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.lfilter.html
# Finally, use plt.plot() to plot the filtered waveform.

# Generate a 10 Hz sine wave
frequency = 10  # Frequency of the sine wave in Hz
duration = 2  # Duration of the sine wave in seconds
sampling_frequency = 1000  # Sampling frequency
t = np.linspace(0, duration, int(sampling_frequency * duration), endpoint=False)
original_signal = np.sin(2 * np.pi * frequency * t)

# Create a low-pass Butterworth filter
order = 6  # Filter order
cutoff_frequency = 5  # Cutoff frequency in Hz
b, a = sig.butter(order, cutoff_frequency, btype='low', fs=sampling_frequency)

# Apply the filter to the sine wave
filtered_signal = sig.lfilter(b, a, original_signal)

# Plot the original and filtered waveforms
plt.figure(figsize=(10, 6))
plt.plot(t, original_signal, label='Original 10 Hz Sine Wave')
plt.plot(t, filtered_signal, label='Filtered Signal (Low-pass 5 Hz)')
plt.title('Original vs. Filtered Sine Wave')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

### Task 4. Create a Python script that applies a band-pass filter to the sine wave generated in task 1. The cutoff frequencies of the band-pass filter should be 4 Hz and 6 Hz. Plot the filtered waveform using the matplotlib library.

```python
# Use a Buttworth filter for your highpass filter. ex. b, a = sig.butter(Order, cutoff, btype = 'highpass', fs = sampleFrequency)
# Order input argument can be between 1-300 (ideally 1-10)
# Make sure to read on how to input two critical frequencies
# Documentation for sig.butter: https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.butter.html
# Use sig.lfilt() to filter the sine wave. ex. y = sig.lfilt(b, a, x)
# Documentation for sig.lfilt(): https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.lfilter.html
# Finally, use plt.plot() to plot the filtered waveform.

# Generate a 10 Hz sine wave
frequency = 10  # Frequency of the sine wave in Hz
duration = 2  # Duration of the sine wave in seconds
sampling_frequency = 44100  # Sampling frequency (Hz)
t = np.linspace(0, duration, int(sampling_frequency * duration), endpoint=False)
original_signal = np.sin(2 * np.pi * frequency * t)

# Create a band-pass Butterworth filter with cutoff frequencies of 4 Hz and 6 Hz
order = 2  # Filter order
cutoff_frequencies = [4, 6]  # Cutoff frequencies in Hz for the band-pass filter
b, a = sig.butter(order, cutoff_frequencies, btype='bandpass', fs=sampling_frequency)

# Apply the filter to the sine wave
filtered_signal = sig.lfilter(b, a, original_signal)

# Plot the original and filtered waveforms
plt.figure(figsize=(10, 4))
plt.plot(t, original_signal, label='Original 10 Hz Sine Wave')
plt.plot(t, filtered_signal, label='Filtered Signal (Band-pass 4-6 Hz)')
plt.title('Original vs. Filtered Sine Wave')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.ylim(-1, 1)  # Set the y-axis range between -1 and 1
plt.show()

```

### Task 5. Create a Python script that performs convolution of the sine wave generated in task 1 with a rectangular window of length 100 samples. Plot the convolved signal using the matplotlib library.

```python
# Suggestion: Create a rectangular window of length 100 samples using the np.ones() function.
# Then use the convolve() function from numpy to perform the convolution.
# Finally, use plt.plot() to plot the convolved signal.

# Generate a 10 Hz sine wave
frequency = 10  # Frequency of the sine wave in Hz
duration = 2  # Duration of the sine wave in seconds
sampling_frequency = 1000  # Sampling frequency (Hz)
t = np.linspace(0, duration, int(sampling_frequency * duration), endpoint=False)
original_signal = np.sin(2 * np.pi * frequency * t)

# Define rectangular window
window_length = 100
rectangular_window = np.ones(window_length)

# Perform the convolution
convolved_signal = np.convolve(original_signal, rectangular_window, mode='same')

# Plot the convolved signal
plt.figure(figsize=(10, 6))
plt.plot(t, original_signal, label='Original 10 Hz Sine Wave')
plt.plot(t, convolved_signal, label='Convolved Signal')
plt.title('Convolved Sine Wave with a Rectangular Window')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

### Task 6. Create a Python script that performs moving average of the sine wave generated in task 1 with a window of length 10 samples. Plot the filtered signal using the matplotlib library.

```python
# Suggestion: Use the numpy function np.convolve() with mode='valid' to perform the moving average.
# Can also be accomplished with for/while loop
# Then use plt.plot() to plot the filtered signal.

# Generate a 10 Hz sine wave
frequency = 10  # Frequency of the sine wave in Hz
duration = 2  # Duration of the sine wave in seconds
sampling_frequency = 44100  # Sampling frequency (Hz)
t = np.linspace(0, duration, int(sampling_frequency * duration), endpoint=False)
sine_wave = np.sin(2 * np.pi * frequency * t)

# Create a window for the moving average of length 10 samples
window_length = 10
moving_average_window = np.ones(window_length) / window_length

# Perform the moving average using convolution
moving_average = np.convolve(sine_wave, moving_average_window, mode='valid')

# The time vector needs to be shortened to match the 'valid' mode output length
t_shortened = t[:len(moving_average)]

# Plot the filtered signal
plt.figure(figsize=(10, 6))
plt.plot(t_shortened, moving_average, label='Moving Average Filtered Signal')
plt.title('Moving Average of Sine Wave')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

### Task 1. Create a Python script that generates a signal that is a sum of two sine waves with different frequencies and amplitudes. The first sine wave should have a frequency of 1000 Hz and an amplitude of 1, the second sine wave should have a frequency of 2000 Hz and an amplitude of 0.5. Plot the waveform using the matplotlib library.

```python
# Suggestions:
# - Use the numpy library's sine function to generate the sine waves
# - Use the numpy library's arange function to create an array of time values
# - Use the matplotlib library's plot function to plot the waveform

import numpy as np
import matplotlib.pyplot as plt

# Generate signal parameters
frequency1 = 10  # Frequency of the first sine wave in Hz
amplitude1 = 1   # Amplitude of the first sine wave
frequency2 = 20  # Frequency of the second sine wave in Hz
amplitude2 = 0.5 # Amplitude of the second sine wave
duration = 1     # Duration of the signal in seconds
sampling_frequency = 1000  # Sampling frequency in Hz

# Create an array of time values
t = np.arange(0, duration, 1/sampling_frequency)

# Generate the sine waves
sine_wave1 = amplitude1 * np.sin(2 * np.pi * frequency1 * t)
sine_wave2 = amplitude2 * np.sin(2 * np.pi * frequency2 * t)

# Sum the two sine waves
combined_signal = sine_wave1 + sine_wave2

# Plot the waveform
plt.figure(figsize=(10, 6))
plt.plot(t, combined_signal, label='Combined Sine Wave')
plt.title('Sum of Two Sine Waves')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

### Task 2. Create a Python script that applies a Butterworth filter to the signal generated in task 1. The cutoff frequency of the filter should be 1500 Hz and the filter order should be 4. Plot the filtered waveform using the matplotlib library.

```python
# Suggestions:
# - Use the scipy library's signal butter function to generate the Butterworth filter
# - Use the scipy library's lfilter function to apply the filter to the signal
# - Use the matplotlib library's plot function to plot the filtered waveform
# - Documentation for sig.butter: https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.butter.html
# - Documentation for sig.lfilt(): https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.lfilter.html

from scipy.signal import butter, lfilter

# Generate a 10 Hz sine wave (from Task 1)
frequency = 10  # Frequency of the sine wave in Hz
amplitude = 1   # Amplitude of the sine wave
duration = 1    # Duration of the signal in seconds
sampling_frequency = 1000  # Sampling frequency in Hz

# Create an array of time values
t = np.arange(0, duration, 1/sampling_frequency)

# Generate the sine wave
signal = amplitude * np.sin(2 * np.pi * frequency * t)

# Butterworth filter parameters
cutoff = 15  # Cutoff frequency in Hz
order = 4    # Filter order

# Generate Butterworth filter coefficients
b, a = butter(order, cutoff / (0.5 * sampling_frequency), btype='low')

# Apply the filter to the signal
filtered_signal = lfilter(b, a, signal)

# Plot the filtered waveform
plt.figure(figsize=(10, 6))
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.title('Butterworth Filtered Waveform')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

### Task 3. Create a Python script that computes the Fast Fourier Transform (FFT) of the signal generated in task 1. Plot the magnitude and phase of the FFT using the matplotlib library.

```python
# Use the fft function to get the Fourier Transform of the sine wave. ex. X = fft(x)
# Documentation for fft: https://docs.scipy.org/doc/scipy/reference/generated/scipy.fft.fft.html
# To get the magnitude of the fft use np.abs(). ex. mag = np.abs(X)
# To get the phase of the fft use np.angle(). ex. phase = np.angle(X)
# Additionally, you can use fftfreq to get the independent axis for your plots. ex. freq = fftfreq(len(X), 1/fs)
# Documentation for fftfreq: https://docs.scipy.org/doc/scipy/reference/generated/scipy.fft.fftfreq.html

# Define the parameters
f1 = 10 # frequency of the first sine wave in Hz
a1 = 1 # amplitude of the first sine wave
f2 = 20 # frequency of the second sine wave in Hz
a2 = 0.5 # amplitude of the second sine wave
t = np.linspace(0, 1, 1000) # time vector in seconds
fs = 1000 # sampling frequency in Hz

# Generate the signals
y1 = a1 * np.sin(2 * np.pi * f1 * t) # first sine wave
y2 = a2 * np.sin(2 * np.pi * f2 * t) # second sine wave
y = y1 + y2 # sum of the two sine waves

# Compute the FFT
yf = fft(y) # complex-valued FFT
xf = fftfreq(len(y), 1/fs) # frequency vector

# Plot the magnitude and phase of the FFT
plt.figure(figsize=(10, 6)) # create a figure
plt.subplot(2, 1, 1) # create a subplot for the magnitude
plt.plot(xf, np.abs(yf)) # plot the magnitude
plt.xlabel('Frequency (Hz)') # label the x-axis
plt.ylabel('Magnitude') # label the y-axis
plt.title('FFT of the sum of two sine waves') # add a title
plt.grid() # add a grid
plt.subplot(2, 1, 2) # create a subplot for the phase
plt.plot(xf, np.angle(yf)) # plot the phase
plt.xlabel('Frequency (Hz)') # label the x-axis
plt.ylabel('Phase (rad)') # label the y-axis
plt.grid() # add a grid
plt.tight_layout() # adjust the spacing
plt.show() # show the plot
```
